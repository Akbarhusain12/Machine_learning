```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
np.random.seed(0)
x = np.linspace(-3,3,20).reshape(-1,1)
y = x ** 2 + np.random.normal(0,1, x.shape)
```

```python
model = LinearRegression()
poly_feat = PolynomialFeatures(degree=5)

poly_model = LinearRegression()

model.fit(x,y)

X_poly = poly_feat.fit_transform(x)

poly_model.fit(X_poly,y)
```

```python
X_test = np.linspace(-3,3,100).reshape(-1,1)
y_linear_pred = model.predict(X_test)
y_poly_pred = poly_model.predict(poly_feat.transform(X_test))
```

```python
plt.scatter(x,y, color="black", label="Actual Data")
plt.plot(X_test,y_linear_pred, label="Underfitting",color="blue")
plt.plot(X_test,y_poly_pred, label="Overfitting D=5",color="red")

plt.title("underfitting vs overfitting (Quadratic Data)")
plt.legend()
plt.show()
```
