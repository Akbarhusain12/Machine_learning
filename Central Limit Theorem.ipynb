```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("Files/data.csv")
```

```python
df['price'].plot(kind="kde")
```

```python
df['price'].sample(100)
```

```python
samples = []

for i in range(100):
    samples.append(df['price'].sample(100).values.tolist())
```

```python
samples
```

```python
samples = np.array(samples)
```

```python
samples
```

```python
sampling_mean = samples.mean(axis = 1)
```

```python
import seaborn as sns
```

```python
sns.kdeplot(sampling_mean)
```

```python
sampling_mean.mean()
```

```python
sampling_mean.std()/np.sqrt(100)
```

```python
lower = sampling_mean.mean() - 2 * sampling_mean.std()/np.sqrt(100)
upper = sampling_mean.mean() + 2 * sampling_mean.std()/np.sqrt(100)
```

```python
print("range", lower, "-", upper)
```
