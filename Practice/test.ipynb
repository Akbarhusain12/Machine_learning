```python
X = df.drop(columns="price")
y = df[['price']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model  = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt

# Scatter plot
plt.scatter(df["sqft_living"], df["price"], edgecolor='k', facecolor='blue', alpha=0.7, label='Sample data')

# Example regression line (just for visualization purposes)
# Let's compute a simple line using two points based on min and max of sqft_living
x1 = df["sqft_living"].min()
x2 = df["sqft_living"].max()

# Example slope and intercept (you can replace these with actual model coefficients)
slope = 300
intercept = 50000

y1 = slope * x1 + intercept
y2 = slope * x2 + intercept

# Plot the regression line using just these two points
plt.plot([x1, x2], [y1, y2], color="red", label="Regression model")

# Labels and Title
plt.xlabel("Size (sq ft)")
plt.ylabel("Price ($)")
plt.legend()
plt.title("House Price Prediction based on Size")

plt.show()
```
