```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('Files/california_housing.csv')
df = pd.DataFrame(data)
df
```

```python
X = df.drop(columns=['MedHouseVal'])
Y = df['MedHouseVal']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,random_state=42)

model_linear = LinearRegression()
model_linear.fit(X_train, Y_train)
Y_pred = model_linear.predict(X_test)
```

```python
# Create a figure and 3D axis
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for actual data
ax.scatter(X_test["MedInc"], X_test["AveRooms"], Y_test, c="blue", label="Actual Data")  # Use 'c' instead of 'color'

# Creating a meshgrid for surface plot
x1_range = np.linspace(X_test["MedInc"].min(), X_test["MedInc"].max(), 100)
x2_range = np.linspace(X_test["AveRooms"].min(), X_test["AveRooms"].max(), 100)
x1, x2 = np.meshgrid(x1_range, x2_range)

# Predicting values for the meshgrid
z = model_linear.predict(np.c_[x1.ravel(), x2.ravel()]).reshape(x1.shape)

# Surface plot
ax.plot_surface(x1, x2, z, facecolors='red', alpha=0.5, rstride=5, cstride=5)  # Use 'facecolors' instead of 'color'

# Labels and title
ax.set_xlabel('Median Income')
ax.set_ylabel('Average Rooms')
ax.set_zlabel('House Price')
ax.set_title('3D Surface Plot using Meshgrid')

plt.legend()
plt.show()
```
